## Model Building :

### 1.Data Preprations:

```python
X = data[data.columns.drop('Chance of Admit ')]
y = data["Chance of Admit "]
```

```python
X.shape
```

```python
y.shape
```

### 2. Train test split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Applying different types of Algorithms

#### 1.Linear Regression

```python
#model
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

```python
model = LinearRegression(normalize=True)
model.fit(X_train, y_train)

print("Mean squared error",mean_squared_error(y_test, model.predict(X_test)))
print ("Variance score : ",model.score(X_test, y_test)*100)
print("Root mean squared error",np.sqrt(mean_squared_error(y_test, model.predict(X_test))))
print("Mean absolute error",mean_absolute_error(y_test, model.predict(X_test)))
print("R2 score",r2_score(y_test,model.predict(X_test)))
methodDict = {}
methodDict['Linear Regression'] = mean_squared_error(y_test, model.predict(X_test))*100
```

```python
from sklearn.ensemble import RandomForestRegressor
rgr=RandomForestRegressor()
rgr.fit(X_train,y_train)
```

```python
print("Mean squared error",mean_squared_error(y_test, rgr.predict(X_test)))
print ("Variance score : ",rgr.score(X_test, y_test)*100)
print("Root mean squared error",np.sqrt(mean_squared_error(y_test, rgr.predict(X_test))))
print("Mean absolute error",mean_absolute_error(y_test, rgr.predict(X_test)))
print("R2 score",r2_score(y_test,rgr.predict(X_test)))
methodDict['Random Forest'] = mean_squared_error(y_test, rgr.predict(X_test))*100
```

### Visualizing the accuracy of the three models

```python
def plotSuccess():
    s = pd.Series(methodDict)
    s = s.sort_values(ascending=False)
    plt.figure(figsize=(12,8))
    #Colors
    ax = s.plot(kind='bar') 
    for p in ax.patches:
        ax.annotate(str(round(p.get_height(),2)), (p.get_x() * 1.005, p.get_height() * 1.005))
    plt.ylim([0, 1])
    plt.xlabel('Method')
    plt.ylabel('MSE x 100')
    plt.title('Success of methods')
     
    plt.show()
plotSuccess()
```

## Linear Regression is the one with the lowest MSE
### We will proceed with Linear Regression model
